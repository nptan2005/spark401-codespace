```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit

spark = SparkSession.builder.appName('Tan').getOrCreate()

df = spark.read.option("header",True).option("inferSchema",True).csv('content/Cricket_data_set_odi.csv')

# Display Schema
df.printSchema()

# Show Dataframe
df.show()

df.withColumn(
    'Avg_runs', df.Runs / df.Matches).withColumn(
    'wkt+10', df.Wickets+10).show()

df.select('*', (df.Runs / df.Matches).alias('Avg_runs'),
          (df.Wickets+10).alias('wkt+10')).show()

df.select('*',lit("Cricket").alias("Sport")). \
    withColumn("Fitness",lit(("Good"))).show()
```
