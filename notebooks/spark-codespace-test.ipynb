```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import when, lit, col
from pyspark.sql.types import *

spark = (
    SparkSession.builder
    .appName("EX")
    .master("local[2]")
    .config("spark.driver.memory", "2g")
    .config("spark.sql.shuffle.partitions", "8")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")

schema = StructType([
    StructField("Name", StringType()),
    StructField("Course_Name", StringType()),
    StructField("Months", IntegerType()),
    StructField("Course_Fees", IntegerType()),
    StructField("Discount", IntegerType()),
    StructField("Start_Date", StringType()),
    StructField("Payment_Done", BooleanType()),
])

data = [
    ("Amit Pathak", "Python", 3, 10000, 1000, "02-07-2021", True),
    ("Shikhar Mishra", "Soft skills", 2, 8000, 800, "07-10-2021", False),
    ("Shivani Suvarna", "Accounting", 6, 15000, 1500, "20-08-2021", True),
    ("Pooja Jain", "Data Science", 12, 60000, 900, "02-12-2021", False),
]

df = spark.createDataFrame(data, schema=schema)

df2 = (
    df
    .withColumn("Status", lit(0))
    .withColumn(
        "Great_Discount",
        when(col("Discount") >= 1000, lit("Yes")).otherwise(lit("No"))
    )
    .withColumn("newCol", lit(1))
)

df2.show()

spark.stop()
```
